now we can go forward to fit a model into this data

```r
y_obs<-data[,2]  #slicing out scores
x_obs<-data[,1]  #slicing out study hours
```

```r
data_train<-data[1:10,]
print("Training Data created")
dim(data_train)
```

```r
fit=lm(data$Scores~data$Hours)        #fitting Y=m*X+c
print("Training Successfull")
fit
```

```r
summary(fit)
```

so we've fitted a model Y=9.776*X+2.484 

we'll now proceed to chceck goodness of the fit

```r
y_fit<-9.776*x_obs+2.484 
```

```r
df<-data.frame(y_fit,y_obs)
df_tab<-table(y_fit,y_obs)
head(df)
dim(df)
```

```r
plot(y_fit,y_obs,main="scatter plot of fitted data against observed data",xlab="fitted value",ylab="observed value")
abline(fit)
```

```r
fit.res=resid(fit)
plot(fit.res,x_obs,main="residuals plot against observed control variable",xlab="Residuals",ylab="control variable")
```

```r
fit.res
```

```r
options(warn=-1)
chisq.test(df_tab)
```

we've successfully fitted a model
